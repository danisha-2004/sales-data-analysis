```python
import matplotlib.pyplot as plt

region=df.groupby("Region")["TotalSales"].sum()

region.plot(kind="bar")

plt.show()
```

```python
month=df.groupby("Month")["TotalSales"].sum()

month.plot(kind="line")

plt.show()
```

```python
cat=df.groupby("Category")["TotalSales"].sum()

cat.plot(kind="pie")

plt.show()
```
